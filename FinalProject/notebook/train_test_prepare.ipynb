```python
from shutil import copy
from collections import defaultdict
import os 
```

```python
def prepare_data(filepath, src, dest):
    class_images = defaultdict(list)
    with open(filepath, 'r') as file:
        paths = [read.strip() for read in file.readlines()]
        for p in paths:
            food = p.split('/')
            class_images[food[0]].append(food[1] + '.jpg')
        
        for food in class_images.keys():
            print(f"\nCopying images into {food}")
            if not os.path.exists(os.path.join(dest, food)):
                os.makedirs(os.path.join(dest, food))
            for image in class_images[food]:
                copy(os.path.join(src, food, image), os.path.join(dest, food, image))
        print("Copied")
```
