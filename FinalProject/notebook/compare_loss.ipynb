```python
import matplotlib.pyplot as plt

origin_clip_loss_path = "/Users/anjonghyeon/Desktop/KU/3-2/DeepLearning/20242R0136COSE47402/FinalProject/src/origin_clip_loss.txt"
sept_clip_loss_path = "/Users/anjonghyeon/Desktop/KU/3-2/DeepLearning/20242R0136COSE47402/FinalProject/src/sept_clip_loss.txt"
```

```python
origin_clip_loss, sept_clip_loss = [], []

with open (origin_clip_loss_path, 'r') as file:
    losses = file.readlines()
    for idx, loss in enumerate(losses):
        if idx % 1000 == 0:
            origin_clip_loss.append(float(loss))
len(origin_clip_loss)
```

```python
with open(sept_clip_loss_path, 'r') as file:
    losses = file.readlines()
    for idx, loss in enumerate(losses):
        if idx % 1000 == 0:
            sept_clip_loss.append(float(loss))
len(sept_clip_loss)
```

```python
indexes = list(range(1, len(origin_clip_loss) + 1))
plt.plot(indexes, origin_clip_loss, marker='o', label='Origin Clip Loss')
plt.plot(indexes, sept_clip_loss, marker='o', label='Sept Clip Loss')

plt.title("Comparison of origin clip, sept clip loss in fine-tuning")
plt.xlabel("iter per 500")
plt.ylabel("loss")
plt.legend()
plt.show()
```
